```python
class InitialiserPageExtractor:
    def __init__(self, driver: Driver, selectors:dict) -> None:
        self.driver = driver
        self.selectors = selectors
        self.data = []
        self.extract()

    def extract(self):
        container = soupify(self.driver.select(bs4_ext.create_selector(self.selectors.get("container"))))
        # print(container)
        datas = bs4_ext.get_all_element_by_locator(container, self.selectors.get("datas"))
        # print(len(datas))
        for data in datas:
            item = {}
            for key in self.selectors.get("data").keys():
                field_selector = self.selectors.get("data").get(key)
                item[key] = bs4_ext.extract_element_by_locator(data, field_selector)
            self.data.append(item)
```

```python
import re
from unidecode import unidecode

def slugify(text: str) -> str:
    # Normalise les caractères (accents, etc.)
    text = unidecode(text)
    # Convertit en minuscules
    text = text.lower()
    # Remplace les caractères non alphanumériques par des tirets
    text = re.sub(r"[^a-z0-9]+", "-", text)
    # Supprime les tirets en tête et queue
    text = text.strip("-")
    # Réduit les tirets multiples en un seul
    text = re.sub(r"-+", "-", text)
    return text

def formated_url_from(url_dict: dict, name: str) -> str:
    """
    url_dict: dictionnaire avec au moins la clé 'url' (ex: "https://www.sudtourisme.nc/offres/.../")
    name: nom de l'établissement
    Retourne: string formatté selon les règles observées.
    """
    # Extraire l'url de base
    url = url_dict.get("url", "")
    # Déterminer le chemin cible
    # Si le nom contient des indications de "gîte" ou "gite", on utilise/hebergement
    lower_name = name.lower()
    use_hebergement = ("gîte" in lower_name) or ("gite" in lower_name) or ("gîte" in lower_name)

    # Déterminer le slug du nom
    slug = slugify(name)

    # Choisir le chemin
    prefix = "/reserver-mon-hebergement/" if use_hebergement else "/reserver-mon-etablissement/"

    # Composer l'URL final
    base = "https://www.book-sudtourisme.com"
    formatted = f"{base}{prefix}{slug}"

    return formatted
```

```python
def get_page_number(driver, selectors) -> int:
    print("Getting page number...")
    # try:
    soupe_element = soupify(driver.select(bs4_ext.create_selector(selectors['pagination'])))
    page_items = bs4_ext.get_all_element_by_locator(soupe_element, selectors["pages"])
    print(page_items[-1])
    page_number = bs4_ext.extract_element_by_locator(page_items[-1], selectors["page"])
    if page_number:
        return int(page_number)
    else:
        print("Page number not found.")
        return
    # except Exception:
    #     print("Failed to get page number.")
```
