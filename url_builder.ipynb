```python
import time
from urllib.parse import parse_qs
from datetime import datetime, timedelta
```

```python
freq = 1
start_date = "01/02/2026"
end_date = "28/02/2026"
```

```python
start_date = datetime.strptime(start_date, "%d/%m/%Y")
end_date = datetime.strptime(end_date, "%d/%m/%Y")
```

```python
stations = [{
        "url": "https://www.sudtourisme.nc/offres/gondwana-city-green-art-noumea-fr-2772109/",
        "name": "Gondwana City Green and Art"
    },
    {
        "url": "https://www.sudtourisme.nc/offres/rivland-resort-paita-fr-2774929/",
        "name": "Rivland Resort"
    },
    {
        "url": "https://www.sudtourisme.nc/offres/hotel-le-lagon-noumea-fr-2772503/",
        "name": "Hotel Le Lagon"
    }]
```

```python
def normalize_url(stations:list) -> list:
        base_url = "https://www.book-sudtourisme.com/reserver-mon-hebergement/"
        urls = []
        for station in stations:
            normalized_url = base_url + station['name'].lower().replace(' ', '-')
            urls.append({"station": station['name'], "url": normalized_url, "detail_url": station['url']})
        return urls
```

```python
def generate_urls(stations:list) -> list:
        urls = []
        date_space = int((end_date - start_date).days) + 1
        checkin = start_date
        checkout = checkin + timedelta(days=freq)  
        if freq in [1, 3, 7]:
            for _ in range(date_space):
                for station in stations:
                    url = station.get('url') + f"?checkin={checkin.strftime('%Y-%m-%d')}&checkout={checkout.strftime('%Y-%m-%d')}"
                    urls.append(url)
                checkin += timedelta(days=1)
                checkout = checkin + timedelta(days=freq)
        return urls
```

```python
normalized_urls = normalize_url(stations)
normalized_urls
```

```python
generated_urls = generate_urls(normalized_urls)
generated_urls
```
