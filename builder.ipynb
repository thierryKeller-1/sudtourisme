recuperer le nombre de pagination

```python
def get_page_number(driver, selectors) -> int:
    print("Getting page number...")
    # try:
    soupe_element = soupify(driver.select(bs4_ext.create_selector(selectors['pagination'])))
    page_items = bs4_ext.get_all_element_by_locator(soupe_element, selectors["pages"])
    print(page_items[-1])
    page_number = bs4_ext.extract_element_by_locator(page_items[-1], selectors["page"])
    if page_number:
        return int(page_number)
    else:
        print("Page number not found.")
        return
    # except Exception:
    #     print("Failed to get page number.")
```
